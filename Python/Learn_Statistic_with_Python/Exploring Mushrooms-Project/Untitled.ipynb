In this project, you will analyze an extensive mushroom dataset from [UCI](https://archive.ics.uci.edu/ml/datasets/Mushroom) using bar charts and acquaint yourself with the diverse array of mushrooms that exist worldwide.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

## Looking the Data

```python
mushroom_data = pd.read_csv('mushroom-data.csv')
mushroom_data.head()
```

```python
mushroom_data.info()
```

## Plotting Bar Graph

```python
columns = mushroom_data.columns.tolist()
print(columns)
```

```python
for column in columns:
    print(column)
```

```python
for column in columns:
    sns.countplot(x=mushroom_data[column], order=mushroom_data[column].value_counts().index, palette='pastel')
    # rotare value labels
    plt.xticks(rotation=30, fontsize=10)
    plt.xlabel(column, fontsize=12)
    plt.title(column + " value Counts")
    plt.show()
    plt.clf()
```

##  Extensions
- Turn any bar graph with less than six bars into a pie chart (hint: use a conditional statement!).

```python
for column in columns:
    if mushroom_data[column].nunique() <= 5:
        continue
    else:
        plt.figure(figsize=(10,5))
        mushroom_data[column].value_counts().plot(kind='pie', autopct='%0.1f%%', shadow=True)
        # rotare value labels
        plt.axis('equal')
        plt.title(column + " value Counts")
        plt.legend(loc='center right', bbox_to_anchor=(1.25, 0.5))
        plt.show()
        plt.clf()     
```

- Create your bar charts using a list comprehension instead of a for loop.

```python
def bar_graph():
    for column in columns:
        sns.countplot(x=mushroom_data[column], order=mushroom_data[column].value_counts().index, palette='pastel')
        # rotare value labels
        plt.xticks(rotation=30, fontsize=10)
        plt.xlabel(column, fontsize=12)
        plt.title(column + " value Counts")
        plt.show()
        plt.clf()

bar_graph()
```

- Remove any graphs you find uninformative.

```python
def bar_graph(data, column):
    sns.countplot(x=data[column], order=data[column].value_counts().index, palette='pastel')
    # rotare value labels
    plt.xticks(rotation=30, fontsize=10)
    plt.xlabel(column, fontsize=12)
    plt.title(column + " value Counts")
    plt.show()
    plt.clf()
```

```python
# class
bar_graph(mushroom_data, 'Class')
```

```python
# population
bar_graph(mushroom_data, 'Population')
```

```python
# habitat
bar_graph(mushroom_data, 'Habitat')
```

```python
# cap surface
bar_graph(mushroom_data, 'Cap Surface')
```
